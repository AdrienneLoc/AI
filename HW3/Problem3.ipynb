```python
def DFS(maze, start, goal):
  stack = []
  visited = set()
  stack.append(start)
  visited.add(start)
  goalReached = False

  while stack:
    #print(stack)
    node = stack.pop()
    #print(node, end=" ")
    if node == goal:     #have we reached the goal?
      goalReached = True
      break;
    for nodeN in maze[node]:    #add all the child nodes into the stack if they're not already there
      if nodeN not in visited:
        visited.add(nodeN)
        stack.append(nodeN)
  if(goalReached):
    print("SUCCESS")
  else:
    print("FAILURE")




maze1 = {
    0: [1, 10],
    1: [0, 2],
    2: [12],
    3: [4, 13],
    4: [3, 5],
    5: [4, 15],
    6: [7, 16],
    7: [6, 8],
    8: [7, 9],
    9: [8, 19],
    10: [0, 20],
    11: [12],
    12: [2, 11],
    13: [3, 14],
    14: [13, 24],
    15: [5, 16],
    16: [6, 15, 26],
    17: [27],
    18: [19, 28],
    19: [9, 18],
    20: [10, 21],
    21: [20, 31],
    22: [23, 32],
    23: [22],
    24: [14, 25, 34],
    25: [24, 35],
    26: [16, 27],
    27: [17, 26],
    28: [18, 38],
    29: [39],
    30: [31, 40],
    31: [21, 30, 32],
    32: [22, 31],
    33: [34, 43],
    34: [24, 33, 44],
    35: [25, 36],
    36: [35],
    37: [38, 47],
    38: [28, 37, 39],
    39: [29, 38],
    40: [30, 41],
    41: [40, 42],
    42: [41, 43],
    43: [33, 42],
    44: [34, 45],
    45: [44],
    46: [47],
    47: [37, 46, 48],
    48: [47, 49],
    49: [48, 50],
    50: [49]
}

maze2 = {
    0: [1, 10],
    1: [0, 2],
    2: [12],
    3: [4, 13],
    4: [3, 5],
    5: [4, 15],
    6: [7, 16],
    7: [6, 8],
    8: [7, 9],
    9: [8, 19],
    10: [0, 20],
    11: [12],
    12: [2, 11],
    13: [3, 14],
    14: [13, 24],
    15: [5, 16],
    16: [6, 15, 26],
    17: [27],
    18: [19, 28],
    19: [9, 18],
    20: [10, 21],
    21: [20],
    22: [23, 32],
    23: [22],
    24: [14, 25, 34],
    25: [24, 35],
    26: [16, 27],
    27: [17, 26],
    28: [18, 38],
    29: [39],
    30: [31, 40],
    31: [21, 30, 32],
    32: [22, 31],
    33: [34, 43],
    34: [24, 33, 44],
    35: [25, 36],
    36: [35],
    37: [38, 47],
    38: [28, 37, 39],
    39: [29, 38],
    40: [30, 41],
    41: [40, 42],
    42: [41, 43],
    43: [33, 42],
    44: [34, 45],
    45: [44],
    46: [47],
    47: [37, 46, 48],
    48: [47, 49],
    49: [48, 50],
    50: [49]
}

DFS(maze1, 0, 50)
DFS(maze2, 0, 50)
```
